### Jednoczynnikowa analiza wariancji

Jednoczynnikowa analiza  wariancji - jest to uogólniony test różnic między średnimi. Można ją rozumieć jako rozszerzenie testu t-Studenta. Test t stosujemy kiedy testujemy różnicę między dwiema średnimi. Natomiast w przypadku kiedy zmienna niezależna przybiera więcej niż 2 poziomy (testujemy różnicę dla więcej niż 2 średnich). Analiza wariancji, podobnie jak w testach t-Studenta, stosujemy zarówno w planach dla grup niezależnych jak i zależnych
  
<br />

W badaniu tym stawiamy dwie hipotezy. W celu ich zrozumienia omówie teraz symbole i oznaczenia:

  * r = liczba populacji
  * $σ^2$ = wariancja
  * $μ$ = wartość oczekiwana
  * $μ_i$ = gdzie $i = 1, ..., r$

<br />

Aby dokonać tego rodzaju analizy wariancji musimy spełnić następujące założenia:

* zmienna zależna mierzona na skali ilościowej
* każda z $r$ populacji musi mieć rozkład normalny
* próby poddawane analizie muszą być od siebie niezależne. Inaczej, założenie o jednorodności wariancji dla analizy wariancji dla grup niezależnych (jednorodność wariancji wyników oznacza, że zróżnicowanie wyników w poszczególnych grupach badawczych jest podobne).
* próby z każdej z populacji muszą być losowymi próbami prostymi. Oznacza to, że każdy element z populacji ma jednakową sznasę no to, aby znaleźć się w próbie
* wariancje w wylosowanych próbach są równe, ponieważ wówczas analiza wariancji jest odporna na zaburzenie większości założeń, np. założenie o kształcie rozkładu czy homogeniczności wariancji.

<br />

Z populacji $r$ losujemy niezależne próby o liczebnościach $n_i$. Na tych próbach dokonujemy obliczeń, których wynikiem są wartości $x_{ij}$ dla $i=1,...,r$ oraz $j=1,...,n_1$. Całkowita wielkość próby to $n=n_1+n_2+...+n_r$.

  W tym momencie stawiamy dwie wcześniej wspomniane hipotezy:

  * hipoteza zerowa: $H_0: μ_1=μ_2=...=μ_r$
  * hipoteza alternatywna: $H_a: μ_1\neqμ_2\ne...\neμ_r$

<br />

Aby zweryfikować powyższe hipotezy obliczamy wartość statystyki F danej wzorem: 

$F=\frac{MSTR}{MSE}$

,gdzie

$MSTR=\frac{1}{r-1}\sum_{i=1}^r{n_i}{(\overline{x_i}-\hat{x})^2}$

$MSE=\frac{1}{n-r}\sum_{i=1}^r\sum_{j=1}^{n_i}{n_i}{({x_{ij}}-\overline{x_i})^2}$

,gdzie

$\overline{x_i}$ = średnia arytmetyczna z i-tej próby

$\hat{x}$ = średnia arytmetyczna ze wszystkich obserwacji ze wszystkich r prób

<br />

Przy założeniu poprawności hipotezy zerowej statystyka F ma rozkład F-Snedecora z $r-1$ stopniami swobody w liczniku i $n-r$ stopniami swobody w mianowniku

<br />

Obszar krytyczny przyjmuje postać

$Q=\{F:F\ge{F_\alpha}\}$

, gdzie

$F_\alpha$ = wartość krytyczna odczytana z tablic rozkładu F-Snedecora dla (r-1, n-r) stopni swobody.

<br />

Warunki przyjęcia bądź przyjęcia/odrzucenia hipotez:

* Jeśli wartość statystyki F należy do obszaru krytycznego Q to odrzucamy hipotezę zerową na rzecz alternatywnej - oznacza to, że badane średnie nie są jednorodne,

* Jeśli wartość statystyki F nie należy do obszary krytycznego Q to nie mamy podstawy do odrzucenia hipotezy zerowej - oznacza to, że badane średnie są jednorodne.

<br />

### Dwuczynnikowa analiza wariancji

Inaczej zwana jako ANOVA dla klasyfikacji podwójnej, bada wpływ dwóch czynników klasyfikujących, podzielonych na wiele poziomów, na wartości badanej cechy mierzalnej. Analiza powinna spełniać następujące założenia: normalność i jednorodność wariancji w próbach wyznaczonych przez czynniki klasyfikujące. W przypadku tej analizy stawiamy 3 hipotezy:

* ${H_{A0}}$ = czynnik A nie różnicuje wyników
* ${H_{B0}}$ = czynnik B nie różnicuje wyników
* ${H_{AB0}}$ = czynniki A i B nie różnicują wyników

### Wczytanie pomiarów czsowych do zmiennych tablicowych

```r
sk700 <- c(2.07, 2.06, 2.08, 2.22, 2.16, 2.14, 2.13, 2.15, 2.13, 2.25, 2.05, 2.08, 2.16, 2.02, 2.06, 1.89, 2.14, 2.08, 2.11, 2.11, 2.00, 1.80, 1.78, 1.80, 1.91, 1.93, 1.92, 1.86, 2.06, 2.01)
sk525 <- c(1.59, 1.77, 1.45, 1.40, 1.51, 1.51, 1.49, 1.39, 1.63, 1.42, 1.29, 1.55, 1.59, 1.49, 1.61, 1.38, 1.55, 1.50, 1.53, 1.60, 1.50, 1.23, 1.41, 1.37, 1.31, 1.40, 1.27, 1.29, 1.49, 1.42)
sk350 <- c(1.06, 1.12, 1.10, 1.16, 1.14, 1.08, 1.21, 1.10, 1.16, 1.05, 1.06, 1.09, 1.15, 1.13, 1.08, 1.24, 1.06, 1.06, 1.08, 1.08, 1.21, 0.98, 1.01, 1.06, 1.11, 0.99, 1.09, 1.11, 0.95, 1.01)
```

### Sprawdzenie normalności danych

```r
shapiro.test(sk700)
shapiro.test(sk525)
shapiro.test(sk350)
```

Po wykonaniu testu Shapiro-Wilka wszystkie wartości p-value są większe od 0.05, stąd można stwierdzić, że mamy do czynienia z rozkładem normalnym. Założenie o normalności danych zostało spełnione.

### Homoskedastyczność danych (równość wariancji)

```r
install.packages("car")
library(car)

pomiary <- data.frame(
  id = c (1:90),
  name = c(rep("70 mm",each=30),rep("52.5 mm", each=30),rep("35 mm", each=30)),
  value = c(sk700,sk525,sk350)
)

pomiary$grupy<-as.factor(pomiary$name)

leveneTest(data=pomiary, value ~ grupy)
```

Pr(>F) jest mniejsze od 0.05, co oznacza, że założenia homoskedastyczności nie zostalo spełnione. Pomimo niespełnienia założenia o homoskedastyczności wykonano analizę wariancji. W przypadku kiedy założenie to nie zostaje spełnione warto skorzystać z analizy Kruskala-Wallisa (testu nieparametrycznego - odpornego na złamanie założeń), której wyniki zamieszczono bezposrednio po analizie waraincji. 

## Analiza wariancji

```r
pomiary_anova <- lm(value ~ name, data = pomiary)
anova(pomiary_anova)
```

Pr(>F) jest znacznie mniejsze od 5% co oznacza, że istnieje statystycznie istotna różnica pomiędzy grupami. Oznacza to, że długość skrzydła helikoptera definitywnie ma znaczenie w długości lotu helikoptera.

### Test Kruskala-Wallisa

Jest to test nieparametryczny, który stosujemy w momencie kiedy złamane są założenia dla jednoczynnikowej wariancji.

```r
kruskal.test(value ~ name, data = pomiary)
```

Wskazuje on na istotne różnice pomiędzy kategoriami, ponieważ wartość p-value jest znacznie mniejsza od 5%.

## Testy post-hoc

Analizy post-hoc służą do przyjrzenia się, które średnie są od siebie różne. Polegają one na porównywaniu każdej średniej w grupie z każdą inną średnią

#### Test Dunnetta

Ponieważ założenie o równości wariancji (homoskedastyczność) nie zostało spełnione to skorzystano z testu Dunnetta, który jest odporny na złamanie tego założenia. 

```r
install.packages("DescTools")
library(DescTools)

set.seed(123)
DunnettTest(pomiary$value, pomiary$grupy)
```

Wykorzystany został test Dunnetta, którego używamy jeżeli mamy nierówne wariancje czyli jeżeli mamy heteroskedastyczność. Wartość p-value tego testu jest znacznie mniejsza od 5%. Wartość p wskazuje na istotne różnice pomiędzy średnimi.

## Testy post-hoc

### Test Tukey'a

Jest testem pierwszego wyboru w przypadku gdy mamy spełnione założenia. Jest podobny do poprawki Boneferroniego, ale ma większą moc przy dużej liczbie porównań, czyli łatwiej wychwycić różnicę pomiędzy średnimi, gdy mamy dużo warunków badawczych.

W przypadku grupy helikopterów o długości skrzydeł 35mm i szerokości spodu 26,5mm nie mamy do czynienia z normalnością, co oznacza, że nie spełniliśmy założenia o normalności danych.

### Homoskedastyczność danych (równość wariancji)

#### Test Bartletta

Test ten stosowany jest w przypadku, gdy mamy do czynienia zarówno z jednokierunkową i dwukierunkową analizą wariancji. W przypadku tego testu musimy spełnić założenie o normalności danych, które nie zostało spełnione w przypadku jednej grupy, ale pomimo tego wykonam ten test.

```r
install.packages("car")
library(car)

pomiary <- data.frame(
  długość = c(rep("35mm",each=30),rep("52.5mm", each=30),rep("70mm", each=30)),
  szerokość = c(rep("17.7mm",each=10),rep("26.5mm", each=10),rep("35mm", each=10),rep("17.7mm",each=10),rep("26.5mm", each=10),rep("35mm", each=10),rep("17.7mm",each=10),rep("26.5mm", each=10),rep("35mm", each=10)),
  value = c(sk350sp177, sk350sp265, sk350sp350, sk525sp177, sk525sp265, sk525sp350, sk700sp177, sk700sp265, sk700sp350)
)

bartlett.test(value ~ interaction(długość, szerokość), data = pomiary)
```

Po wykonaniu testu Bartletta otrzymano wartość p-value równą 0.2031, co jest wynikiem większym od 5%. Oznacza to, że mamy do czynienia z równością wariancji. Założenie o równości wariancji zostało spełnione.

Po wykonaniu analizy wariancji bez interakcji można zauważyć, że wartości p-value zarówno dla długości skrzydeł jak i szerokości spodu dały wynik (2e-16 oraz 5,33e-9) poniżej 5% co oznacza, że mamy do czynienia z różnicami na poziomie istotnym statystycznie. Wartość p-value jest znacznie mniejsza dla długości skrzydeł co może świadczyć o silniejszych różnicach średnich niż w przypadku szerokości spodu helikopterów.
